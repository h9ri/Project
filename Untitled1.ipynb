```python
def get_2D_dct(img):
    return fftpack.dct(fftpack.dct(img.T, norm='ortho').T, norm='ortho')
```

```python
a=np.zeros((60,60))
for i in range(20):
    a=a+mean_within_class[i]
total_mean_between_class=a/20
```

```python
final_label=[]
for j in image_feature:
    distances=[]
    label=[]
    for i in feature_vector:
        distances.append(np.linalg.norm(i-j))
    abc=np.argsort(distances)[:5]
    for k in abc:
        label.append((k/5)+1)
    label=np.bincount(label)
    final_label.append(np.argmax(label))
```
